```python
import os, shutil
```

```python
path = r"C:/Users/KnownIGL/Desktop/File Sorter Project/"
```

```python
os.listdir(path)
```

```python
file_name = os.listdir(path)
```

```python
folder_names = ['csv files', 'image files', 'docx file']

for loop in range (0,3):
    if not os.path.exists(path + folder_names[loop]):
        print(path + folder_names[loop])
        os.makedirs((path + folder_names[loop]))

for file in file_name:
    if ".csv" in file and not os.path.exists(path + "csv files/" + file):
        shutil.move(path + file, path + "csv files/" + file)
    elif ".png" in file and not os.path.exists(path + "image files/" + file):
        shutil.move(path + file, path + "image files/" + file)
    elif ".jpg" in file and not os.path.exists(path + "image files/" + file):
        shutil.move(path + file, path + "image files/" + file)
    elif ".docx" in file and not os.path.exists(path + "docx file/" + file):
        shutil.move(path + file, path + "docx file/" + file)
    else:
        print('There are files that are of not specified type')
```
